## Media

```python
def media(M):
    row, col = M.shape
    x = np.zeros((row,col))
    
    for i in range(1,row-1):
        for j in range(1,col-1):
            p = M[i-1:i+2,j-1:j+2]#filas:columnas
            s = np.mean(p)
            x[i,j] = s
    return x.astype('uint8')
```

```python
m = media(cv_gray)
m.shape
```

```python
m_gray_display = cv2.cvtColor(m,cv2.COLOR_GRAY2BGR)

plt.imshow(m_gray_display)
plt.axis('off')
plt.show()
```

# Mediana

```python
def mediana(M):
    row, col = M.shape
    x = np.zeros((row,col))
    
    for i in range(1,row-1):
        for j in range(1,col-1):
            p = M[i-1:i+2,j-1:j+2]#filas:columnas
            s = np.median(p)
            x[i,j] = s
    return x.astype('uint8')
```

```python
med = mediana(cv_gray)
med.shape
```

```python
med_gray_display = cv2.cvtColor(med,cv2.COLOR_GRAY2BGR)

plt.imshow(med_gray_display)
plt.axis('off')
plt.show()
```

# Moda

```python
from scipy import stats
```

```python
def moda(M):
    row, col = M.shape
    x = np.zeros((row,col))
    
    for i in range(1,row-1):
        for j in range(1,col-1):
            p = M[i-1:i+2,j-1:j+2]#filas:columnas
            s = stats.mode(p.ravel())[0][0]
            x[i,j] = s
    return x.astype('uint8')
```

```python
mod = moda(cv_gray)
mod.shape
```

```python
mod_gray_display = cv2.cvtColor(mod,cv2.COLOR_GRAY2BGR)

plt.imshow(mod_gray_display)
plt.axis('off')
plt.show()
```

# Min

```python
def minimo(M):
    row, col = M.shape
    x = np.zeros((row,col))
    
    for i in range(1,row-1):
        for j in range(1,col-1):
            p = M[i-1:i+2,j-1:j+2]#filas:columnas
            s = np.min(p.ravel())
            x[i,j] = s
    return x.astype('uint8')
```

```python
minimo = moda(cv_gray)
minimo.shape
```
