## Media

```python
def media(M):
    row, col = M.shape
    x = np.zeros((row,col))
    
    for i in range(1,row-1):
        for j in range(1,col-1):
            p = M[i-1:i+2,j-1:j+2]#filas:columnas
            s = np.mean(p)
            x[i,j] = s
    return x.astype('uint8')
```

```python
m = media(cv_gray)
m.shape
```

```python
m_gray_display = cv2.cvtColor(m,cv2.COLOR_GRAY2BGR)

plt.imshow(m_gray_display)
plt.axis('off')
plt.show()
```

# Mediana

```python
def mediana(M):
    row, col = M.shape
    x = np.zeros((row,col))
    
    for i in range(1,row-1):
        for j in range(1,col-1):
            p = M[i-1:i+2,j-1:j+2]#filas:columnas
            s = np.median(p)
            x[i,j] = s
    return x.astype('uint8')
```

# Moda

```python
from scipy import stats
```

```python
def moda(M):
    row, col = M.shape
    x = np.zeros((row,col))
    
    for i in range(1,row-1):
        for j in range(1,col-1):
            p = M[i-1:i+2,j-1:j+2]#filas:columnas
            s = stats.mode(p.ravel())[0][0]
            x[i,j] = s
    return x.astype('uint8')
```

```python
mod = moda(cv_gray)
mod.shape
```

```python
mod_gray_display = cv2.cvtColor(mod,cv2.COLOR_GRAY2BGR)

plt.imshow(mod_gray_display)
plt.axis('off')
plt.show()
```

# Maximo

```python
def maximo(M):
    row, col = M.shape
    x = np.zeros((row,col))
    
    for i in range(1,row-1):
        for j in range(1,col-1):
            p = M[i-1:i+2,j-1:j+2]#filas:columnas
            s = np.max(p)
            x[i,j] = s
    return x.astype('uint8')
```

```python
maxm = maximo(cv_gray)
maxm.shape
```

```python
maxm_gray_display = cv2.cvtColor(maxm,cv2.COLOR_GRAY2BGR)

plt.imshow(maxm_gray_display)
plt.axis('off')
plt.show()
```

```python
def PuntoMedio(M):
    row, col = M.shape
    x = np.zeros((row,col))
    
    for i in range(1,row-1):
        for j in range(1,col-1):
            p = M[i-1:i+2,j-1:j+2]#filas:columnas
            mini = np.min(p)
            maxi = np.max(p)
            x[i,j] = (mini+maxi)/2
    return x.astype('uint8')
```

```python
puntomedio = PuntoMedio(cv_gray)
puntomedio.shape
```

```python
puntomedio_gray_display = cv2.cvtColor(puntomedio,cv2.COLOR_GRAY2BGR)

plt.imshow(puntomedio_gray_display)
plt.axis('off')
plt.show()
```

# Derivada

```python
x = np.array([[-1, 0, 1], [-2, 0, 2],[-1,0,1]])
x
```

```python
def derivada(M,X):
    row, col = M.shape
    x = np.zeros((row,col), dtype= "uint8")
    #X
    for i in range(1,row-1):
        for j in range(1,col-1):
            p = M[i-1:i+2,j-1:j+2]
            s = np.abs(np.sum(np.multiply(p,X)))#multiplicacion de matrices
            val = min(s,255)
            x[i,j] = val
    return x.astype('uint8')
```

```python
defX = derivada(cv_gray,x)
defX.shape
```

```python
defX_gray_display = cv2.cvtColor(defX,cv2.COLOR_GRAY2BGR)

plt.imshow(defX_gray_display)
plt.axis('off')
plt.show()
```

```python
y = np.array([[-1, -2, -1], [0, 0, 0],[1,2,1]])
y
```

```python
defY = derivada(cv_gray,y)
defY.shape
```

```python
defY_gray_display = cv2.cvtColor(defY,cv2.COLOR_GRAY2BGR)

plt.imshow(defY_gray_display)
plt.axis('off')
plt.show()
```

```python
#suma 
suma = (defX.astype(int) + defY.astype(int))/2
suma = suma.astype("uint8")
suma.shape
```

```python
suma_gray_display = cv2.cvtColor(suma,cv2.COLOR_GRAY2BGR)

plt.imshow(suma_gray_display)
plt.axis('off')
plt.show()
```
