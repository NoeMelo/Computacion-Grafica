```python
def derivada(M,X):
    row, col = M.shape
    x = np.zeros((row,col), dtype= "uint8")
    #X
    for i in range(1,row-1):
        for j in range(1,col-1):
            p = M[i-1:i+2,j-1:j+2]
            s = np.abs(np.sum(np.multiply(p,X)))#multiplicacion de matrices
            val = min(s,255)
            x[i,j] = val
    return x.astype('uint8')
```

```python
x = np.array([[-1, 0, 1], [-2, 0, 2],[-1,0,1]])
x
```

```python
y = np.array([[-1, -2, -1], [0, 0, 0],[1,2,1]])
y
```

```python
path = "C:/Users/laboratorioesan/Desktop/faces/derivada"
for i in grays[3:]:
    img = cv2.imread("C:/Users/laboratorioesan/Desktop/faces/grey/"+i)
    img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    defX = derivada(img,x)
    defY = derivada(img,y)
    suma = (defX.astype(int) + defY.astype(int))/2
    suma = suma.astype("uint8")
    #suma_gray_display = cv2.cvtColor(suma,cv2.COLOR_GRAY2BGR)
    cv2.imwrite(os.path.join(path , i), suma)
    cv2.waitKey(0)
```
