## Vectores Caracteristicos

```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np
import os
```

```python
imagenes = os.listdir('images')
len(imagenes)
```

```python
os.getcwd()
```

```python
def extraeVector(image):
    img = cv2.imread(image)
    cv_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    return cv_gray.ravel()
```

```python
dic = {}
for image in imagenes:
    vec = extraeVector('/home/noe/Documentos/Computacion-Grafica/6. Extraccion de Vectores'+image)
    dic[image] = list(vec)
```
