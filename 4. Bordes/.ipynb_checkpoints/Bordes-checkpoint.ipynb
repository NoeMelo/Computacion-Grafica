```python
%matplotlib inline
from matplotlib import pyplot as plt

import cv2
import numpy as np
```

```python
img = "frutas.png"
imagen = cv2.imread(img)
imagen.shape 
```

```python
plt.imshow(imagen)##imagen en BGR
plt.axis('off')
plt.show()
```

```python
imagen2_RGB = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
plt.imshow(imagen2_RGB)##imagen en RGB
plt.axis('off')
plt.show()
```

```python
imagen2_GRAY = cv2.cvtColor(imagen2_RGB,cv2.COLOR_BGR2GRAY)
plt.imshow(imagen2_GRAY)##imagen en RGB
plt.axis('off')
plt.show()
```

```python
### se puede obtener una de las 3 capas para trabajar porque los 3 niveles en grises son el mismo. 
```

```python
def getPromedio(imagen):
    B = imagen[:,:,0]
    G = imagen[:,:,1]
    R = imagen[:,:,2]
    #B.shape #
    B = B.astype(int)# PASATE A TIPO ENTERO
    G = G.astype(int)
    R = R.astype(int)
    promedio = (B+G+R)/3
    promedio = promedio.astype('uint8') ##
    return promedio
```

```python
promedio = getPromedio(imagen)
promedio.shape
```

```python
row, col = promedio.shape
umbral = np.zeros((row,col))

for i in range(row):
    for j in range(col):
        if promedio[i,j] > 125:
            umbral[i,j] = 255
        else:
            umbral[i,j] = 0
```

```python
umbral = umbral.astype('uint8') ##
umbral.shape
```

```python
umbral_gray = cv2.cvtColor(umbral, cv2.COLOR_GRAY2BGR)
plt.imshow(umbral_gray)
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(17, 8))
plt.subplot(1, 2, 1)
plt.imshow(imagen2_RGB)
plt.title("imagen original")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(umbral_gray)
plt.title("imagen binaria")
plt.axis('off')
```
