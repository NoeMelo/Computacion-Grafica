```python
B = imagen[:,:,0]
G = imagen[:,:,1]
R = imagen[:,:,2]
#B.shape #
B = B.astype(int)# PASATE A TIPO ENTERO
G = G.astype(int)
R = R.astype(int)
```

# Operador Inverso o Negativo
q(x,y) = 255 - p(x,y)

```python
promedio.shape
```

```python
negativo = 255 - promedio
```

```python
negativo_gray = cv2.cvtColor(negativo, cv2.COLOR_GRAY2BGR)
plt.imshow(negativo_gray)
plt.axis('off')
plt.show()
```

# OPerador Umbral
q(x,y) = si <= u entonces 0, si > u entonces 255

```python
row, col = promedio.shape
```

```python
umbral = np.zeros((row,col))
umbral.shape
```

```python
for i in range(row):
    for j in range(col):
        if promedio[i,j] <= 40:
            umbral[i,j] = 0
        else:
            umbral[i,j] = 255
```

```python
umbral=umbral.astype('uint8')
```

```python
umbral_gray = cv2.cvtColor(umbral, cv2.COLOR_GRAY2BGR)
plt.imshow(umbral_gray)
plt.axis('off')
plt.show()
```

# Operador Umbral binario
q(x,y) = 
para p(x,y)  <= u o u2  <= p(x,y) entonces 0, 
para u < p(x,y) < u2 entonces 255

```python
umbralB = np.zeros((row,col))
umbralB.shape
```

```python
for i in range(row):
    for j in range(col):
        if promedio[i,j] <= 40 or 120 <= promedio[i,j]:
            umbralB[i,j] = 0
        elif 40 < promedio[i,j] < 120:
            umbralB[i,j] = 255
```

```python
umbralB=umbralB.astype('uint8')
```

```python
umbralB_gray = cv2.cvtColor(umbralB, cv2.COLOR_GRAY2BGR)
plt.imshow(umbralB_gray)
plt.axis('off')
plt.show()
```
