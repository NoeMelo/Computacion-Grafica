# Escala de Grises

**Generar una matriz promedio**

```python
promedio = np.zeros((len(R),len(R)))
promedio.shape
```

```python
promedio[1,1]
```

```python
for i in range(0, len(R)):
    for j in range(0, len(R)):
        promedio[i,j] = (B[i,j]+G[i,j]+R[i,j])/3      
```

```python
promedio
```

**To integer**

```python
B = B.astype(int)# PASATE A TIPO ENTERO
G = G.astype(int)
R = R.astype(int)
```

```python
promedio = (B+G+R)/3
```

```python
promedio
```

```python
promedio = promedio.astype('uint8') ##
```

```python
promedio_gray = cv2.cvtColor(promedio, cv2.COLOR_GRAY2BGR)
```

```python
plt.imshow(promedio_gray)
plt.axis('off')
plt.show()
```

# Sobreponer dos imagenes - Tarea

```python
keiko = cv2.imread("Keiko1.PNG")
toledo = cv2.imread("Toledo1.PNG")

print(keiko.shape)
print(toledo.shape)
```

```python
keiko = keiko[0:400,0:600]#posicion
plt.imshow(keiko)##imagen en BGR
plt.axis('off')
plt.show()
```

```python
toledo = toledo[0:400,0:600]#posicion
plt.imshow(toledo)##imagen en BGR
plt.axis('off')
plt.show()
```

```python
kB = keiko[:,:,0]
kG = keiko[:,:,1]
kR = keiko[:,:,2]
kB.shape #
```

```python
kB = kB.astype(int)# PASATE A TIPO ENTERO
kG = kG.astype(int)
kR = kR.astype(int)
```

```python
tB = toledo[:,:,0]
tG = toledo[:,:,1]
tR = toledo[:,:,2]
tB.shape #
```

```python
tB = tB.astype(int)# PASATE A TIPO ENTERO
tG = tG.astype(int)
tR = tR.astype(int)
```

```python
cross = (kB+kG+kR + tB+tG+tR)/6
```

```python
cross = cross.astype('uint8') ##
```

```python
cross_gray = cv2.cvtColor(cross, cv2.COLOR_GRAY2BGR)
```

```python
plt.imshow(cross_gray)
plt.axis('off')
plt.show()
```

```python
plt.imshow(cross)
plt.axis('off')
plt.show()
```

```python
cross_RGB = cv2.cvtColor(cross,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(cross_RGB)
plt.axis('off')
plt.show()
```
